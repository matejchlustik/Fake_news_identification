```python
tokenizer = TreebankWordTokenizer()


def preprocess_word(word):
    word = re.sub(r"[‘’]", "'", word)
    return re.sub(r"[“”`]", "", word)


def tokenize_article(article):

    sentences = sent_tokenize(article)

    tokenized_sentences = [
        [preprocess_word(word) for word in tokenizer.tokenize(sentence)]
        for sentence in sentences
    ]
    return [[word for word in sentence if word] for sentence in tokenized_sentences]
```

```python
unique_df["tokenized_text"] = unique_df["text"].apply(tokenize_article)
```
